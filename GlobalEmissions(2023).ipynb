```python
import pandas as pd
```

```python
# Load the real global emissions file from the web:
url = "https://nyc3.digitaloceanspaces.com/owid-public/data/co2/owid-co2-data.csv"
df = pd.read_csv(url)
```

```python
# Step 1 — filter to the year 2023:
df_2023 = df[df["year"] == 2023]
```

```python
# Some rows are whole regions ("World", "Europe"), not real countries.
# Real countries have an ISO code; regions don't. Keep only real countries:
countries = df_2023[df_2023["iso_code"].notna()]
```

```python
# Step 2 — sort by co2 (biggest first) and take the top 50:
top50 = countries.sort_values("co2", ascending=False).head(50)
```

```python
# Show just the two columns we care about:
print(top50[["country", "co2"]])
```
